## Imputing missing values

```python
# Imputing recovered column
df_recovered = pd.DataFrame()
df_recovered['Recovered'] = df['Confirmed']-df['Active']-df['Deaths']
df_recovered
```

```python
df = df.drop(columns=['Recovered'])
df = pd.concat([df_recovered,df],axis=1)
```

```python
# Replacing missing values in mortality rate by 0
df['Mortality_Rate'] = df['Mortality_Rate'].fillna(0)
```

```python
# Dropping columns
df = df.drop(columns=['Lat','Long_'])
```

```python
# Imputing missing values with mean
for col in ['Incident_Rate','Testing_Rate','Hospitalization_Rate']:
    df[col].fillna(df[col].mean(),inplace=True)
```

```python
# Imputing missing values with mode
for col in ['People_Tested','People_Hospitalized']:
    df[col].fillna(df[col].mode()[0],inplace=True)
```
