# Model Training

```python
# seprating the train dataset into input and output variables
x = df.drop(columns=['COST'],axis=1)
y = df['COST']
x.head(10)
```

### Applying PCA to input variables

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=20)
```

```python
x = pca.fit_transform(x)
x.shape
```

```python
df_test = pca.fit_transform(df_test)
df_test.shape
```

```python
# scaling the input variable
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x=sc.fit_transform(x)
```

```python
# Dividing input and target variables into train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=42)
```

```python
# Importing our model Library
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso,Ridge,ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR


# Importing GridSearchCV
from sklearn.model_selection import GridSearchCV, cross_val_score

# Importing Evaluation metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

## Finding Best Parameters

```python
# Using GridSearchCV to find best parameters in Lasso regression model
model_l = Lasso()
alphavalue = {'alpha':[0.1,0.01,0.001,0.0001]}
gd_l = GridSearchCV(estimator=model_l,param_grid=alphavalue,scoring='r2',cv=10)
gd_l.fit(x,y)
best_parameters = gd_l.best_params_
print(best_parameters)
best_result = gd_l.best_score_
print(best_result)
```

```python
# Using GridSearchCV to find best parameters in Ridge regression model
model_r = Ridge()
alphavalue = {'alpha':[0.1,0.01,0.001,0.0001]}
gd_r = GridSearchCV(estimator=model_r,param_grid=alphavalue,scoring='r2',cv=10)
gd_r.fit(x,y)
best_parameters = gd_r.best_params_
print(best_parameters)
best_result = gd_r.best_score_
print(best_result)
```

```python
# Using GridSearchCV to find best parameters in ElasticNet regression model
model_e = ElasticNet()
alphavalue = {'alpha':[0.1,0.01,0.001,0.0001]}
gd_e = GridSearchCV(estimator=model_e,param_grid=alphavalue,scoring='r2',cv=10)
gd_e.fit(x,y)
best_parameters = gd_e.best_params_
print(best_parameters)
best_result = gd_e.best_score_
print(best_result)
```

```python
# GridSearchCV for Decision Tree Regressor
dtr = DecisionTreeRegressor()
grid_param = { 'criterion' : ['mse', 'friedman_mse', 'mae']}
gd_dtr = GridSearchCV(estimator=dtr,param_grid=grid_param,scoring='r2',cv=5)
gd_dtr.fit(x_train,y_train)
best_parameters = gd_dtr.best_params_
print(best_parameters)
best_result = gd_dtr.best_score_
print(best_result)
```

```python
# Models with its best parameters
model_Lasso = Lasso(alpha=0.0001)
model_Ridge = Ridge(alpha=0.1)
EN = ElasticNet(alpha=0.0001)
DTR = DecisionTreeRegressor(criterion='mse')
```

```python
# Using algorithm via for loop
model=[LinearRegression(),SVR(),model_Lasso,model_Ridge,EN,DTR]
for m in model:
    m.fit(x_train,y_train)
    print('Score of ',m,' is: ',m.score(x_train,y_train))
    pred=m.predict(x_test)
    print('ERROR:')
    print('Mean absolute error : ',mean_absolute_error(y_test,pred))
    print('Mean squared error : ',mean_squared_error(y_test,pred))
    print('Root mean squared error : ',np.sqrt(mean_squared_error(y_test,pred)))
    print('R2 score : ',r2_score(y_test,pred))
    print('*********************************************************************')
    print('\n\n')
```

```python
# cross validate the models
models = [LinearRegression(),SVR(),model_Lasso,model_Ridge,EN,DTR]
for m in models:
    score=cross_val_score(m,x,y,cv=10,scoring='r2')
    print('Model : ',m)
    print('Score : ',score)
    print('Mean score : ',score.mean())
    print('Standard deviation : ',score.std())
    print('**************************************************************')
    print('\n\n')
```

## Using Ensemble Models

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import ExtraTreesRegressor
```

```python
# cross validate the models
models = [RandomForestRegressor(),AdaBoostRegressor(),GradientBoostingRegressor(),BaggingRegressor(),ExtraTreesRegressor()]
for m in models:
    score=cross_val_score(m,x,y,cv=10,scoring='r2')
    print('Model : ',m)
    print('Score : ',score)
    print('Mean score : ',score.mean())
    print('Standard deviation : ',score.std())
    print('**************************************************************')
    print('\n\n')
```

## Hyperparameter Tuning For Models With Best Results

```python
# RandomizedSearchCV for Random Forest model
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
random_grid = {'n_estimators': [100,300,500,700],
               'min_samples_split':[2,4,6,8,10],
               'min_samples_leaf':[1,2,3,4],
               'max_depth':[5,10,15,20,25],
               'max_features': ['auto','sqrt','log2']}
model_rf = RandomForestRegressor()
grid_rf = RandomizedSearchCV(estimator=model_rf,
                           param_distributions=random_grid,
                           n_iter = 100, cv = 5, 
                           verbose=2, n_jobs = -1,
                           random_state=42)
grid_rf.fit(x,y)
grid_rf.best_params_
```

```python
# Using GridSearchCV to find best parameters in SVR
svr = SVR()
parameters={'kernel':['rbf','poly','linear','sigmoid'],}
gd_svr = GridSearchCV(estimator=svr,param_grid=parameters,scoring='r2',cv=5)
gd_svr.fit(x,y)
gd_svr.best_params_
```

```python
# Best Models with Best Parameters
model_svr = SVR(kernel = 'rbf')
RF = RandomForestRegressor(n_estimators = 100,
                           min_samples_split = 10,
                           min_samples_leaf= 2,
                           max_features = 'sqrt',
                           max_depth = 15)
GB = GradientBoostingRegressor(n_estimators = 500)
```

```python
# Using algorithm via for loop
model=[model_svr,RF,GB]
for m in model:
    m.fit(x_train,y_train)
    print('Score of ',m,' is: ',m.score(x_train,y_train))
    pred=m.predict(x_test)
    print('ERROR:')
    print('Mean absolute error : ',mean_absolute_error(y_test,pred))
    print('Mean squared error : ',mean_squared_error(y_test,pred))
    print('Root mean squared error : ',np.sqrt(mean_squared_error(y_test,pred)))
    print('R2 score : ',r2_score(y_test,pred))
    print('*********************************************************************')
    print('\n\n')
```

```python
# cross validate the models
models = [model_svr,RF,GB]
for m in models:
    score=cross_val_score(m,x,y,cv=10,scoring='r2')
    print('Model : ',m)
    print('Score : ',score)
    print('Mean score : ',score.mean())
    print('Standard deviation : ',score.std())
    print('**************************************************************')
    print('\n\n')
```
