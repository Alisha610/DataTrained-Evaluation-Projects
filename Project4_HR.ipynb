### Treating Outliers

```python
column_out = ['MonthlyIncome']
for col in column_out:
    q1 = df[col].quantile(0.10)
    q2 = df[col].quantile(0.90)
    print('For ',col,': ',q1,'\t',q2)
    df[col]=np.where(df[col]<q1,q1,df[col])
    df[col]=np.where(df[col]>q2,q2,df[col])
    print(df[col].skew())
    print('\n\n')
```

# Model Training

```python
# seprating the dataset into input and output variables
x = df_new.drop(columns=['Attrition'])
y = pd.DataFrame(df_new['Attrition'])
```

```python
x.dtypes
```

```python
y
```

```python
# Scaling the input variable
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,5))
columns = list(x)
for col in columns:
    x[col]=x[col].astype(float)
    x[[col]]=scaler.fit_transform(x[[col]])
```

```python
x
```

```python
x.shape
```

```python
# Applying pca to the dataset
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
x = pca.fit_transform(x)
x.shape
```

```python
x=pd.DataFrame(x)
x
```

```python
# Dividing input and target variables into train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=42)
```

```python
print(x_train.shape,'\t',y_train.shape)
```

```python
# Importing our model Library
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

# Importing GridSearchCV
from sklearn.model_selection import GridSearchCV

# Importing Evaluation metrics
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
# Using algorithms via for loop
model=[LogisticRegression(),GaussianNB(),DecisionTreeClassifier(),SVC(),KNeighborsClassifier()]
for m in model:
    m.fit(x_train,y_train)
    print('Score of ',m,' is: ',m.score(x_train,y_train))
    pred=m.predict(x_test)
    print('Accuracy score : ',accuracy_score(y_test,pred))
    print('Confusion matrix : \n',confusion_matrix(y_test,pred))
    print('Classification report : \n',classification_report(y_test,pred))
    print('*********************************************************************')
    print('\n\n')
```

SVC model and Logistic Regresssion model are giving the best accuracy scores

```python
# SVC model is giving best result so finding its best parameter using RandomizedSearchCV
from sklearn.model_selection import RandomizedSearchCV
svc = SVC()
parameters={'kernel':['rbf','poly','linear','sigmoid'],
           'C' : [1,5,10],
           'degree' : [3,8],
           'gamma' : ('auto','scale')}
clf = RandomizedSearchCV(estimator=svc,
                           param_distributions=parameters,
                           n_iter = 100, 
                           cv = 5, verbose=2, 
                           n_jobs = -1,
                           random_state=42)
clf.fit(x,y)
clf.best_params_
```

```python
# Using SVR with best parameters
model_svc = SVC(kernel='rbf',C=1,degree=3,gamma='scale')
model_svc.fit(x_train,y_train)
print('Score : ',model_svc.score(x_train,y_train))
y_pred=model_svc.predict(x_test)
print('Accuracy score : ',accuracy_score(y_test,y_pred))
print('Confusion matrix : \n',confusion_matrix(y_test,y_pred))
print('Classification report : \n',classification_report(y_test,y_pred))
```

```python
# Cross validation 
from sklearn.model_selection import cross_val_score
score=cross_val_score(model_svc,x,y,cv=5,scoring='accuracy')
print('Mean score: ',score.mean())
print('Standard deviation : ',score.std())
```

```python
# Logistic Regression is giving best result so finding its best parameter using RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
parameter={'penalty':['l1', 'l2', 'elasticnet'],
          'random_state':range(40,100),
          'solver':['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
           'C':[100,10,1.0,0.1,0.01,0.001]}
clf = RandomizedSearchCV(estimator=lr,
                           param_distributions=parameter,
                           n_iter = 100, 
                           cv = 5, verbose=2, 
                           n_jobs = -1,
                           random_state=42)
clf.fit(x,y)
clf.best_params_
```

```python
# Using Logistic Regression with best parameters
regr = LogisticRegression(penalty='l1',random_state=69,solver='liblinear',C=10)
regr.fit(x_train,y_train)
print('Score : ',regr.score(x_train,y_train))
predregr = regr.predict(x_test)
print('Accuracy score : ',accuracy_score(y_test,predregr))
print('Confusion matrix : \n',confusion_matrix(y_test,predregr))
print('Classification report : \n',classification_report(y_test,predregr))
```

```python
# Cross validation 
from sklearn.model_selection import cross_val_score
score=cross_val_score(regr,x,y,cv=5,scoring='accuracy')
print('Mean score: ',score.mean())
print('Standard deviation : ',score.std())
```

Logistic Regression model gives a better cross validation score and it shows less overfitting of data as compared to SVC model

### Using Ensemble model

```python
# Using RandomizedSearchCV for RandomForest
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
rf = RandomForestClassifier()
parameter={'n_estimators':[50,100,150,200,250],
          'min_samples_split':[2,4,6,8,10],
          'min_samples_leaf':[1,2,3,4],
          'max_depth':[5,10,15,20,25]}
grid_rf = RandomizedSearchCV(estimator=rf,
                           param_distributions=parameter,
                           n_iter = 100, 
                           cv = 5, verbose=2, 
                           n_jobs = -1,
                           random_state=42)
grid_rf.fit(x,y)
grid_rf.best_params_
```

```python
# Using Random Forest with best parameters
rforest = RandomForestClassifier(n_estimators=200,min_samples_split=8,min_samples_leaf=3,max_depth=15)
rforest.fit(x_train,y_train)
print('Score : ',rforest.score(x_train,y_train))
predforest = rforest.predict(x_test)
print('Accuracy score : ',accuracy_score(y_test,predforest))
print('Confusion matrix : \n',confusion_matrix(y_test,predforest))
print('Classification report : \n',classification_report(y_test,predforest))
```

```python
# Cross validation 
score=cross_val_score(rforest,x,y,cv=10,scoring='accuracy')
print('Mean score: ',score.mean())
print('Standard deviation : ',score.std())
```

RandomForestClassifier gives better accuracy, but LogisticRegression model gives less over fitting

### AUC ROC Curve

```python
# Importing the metrics
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
# ROC Curve for Logistic regression model
y_pred_prob = regr.predict_proba(x_test)[:,1]
print('roc_auc_score for Logistic regression model: ', roc_auc_score(y_test, y_pred_prob))
fpr,tpr,threshold = roc_curve(y_test,y_pred_prob)
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr,label='Logistic Regression')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Logistic Regression')
plt.show()
```
